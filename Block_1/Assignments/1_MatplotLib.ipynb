## Exercise 3
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).


### The result should look something like this:
<img src="IMG/3.png" width=400>

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(data.month_number, data.facecream, color="blue", lw=3, ls='-', marker='o', markersize=6, markerfacecolor="blue")
ax.plot(data.month_number, data.facewash, color="orange", lw=3, ls='-', marker='o', markersize=6, markerfacecolor="orange")
ax.plot(data.month_number, data.toothpaste, color="green", lw=3, ls='-', marker='o', markersize=6, markerfacecolor="green")
ax.plot(data.month_number, data.bathingsoap, color="red", lw=3, ls='-', marker='o', markersize=6, markerfacecolor="red")
ax.plot(data.month_number, data.shampoo, color="purple", lw=3, ls='-', marker='o', markersize=6, markerfacecolor="purple")
ax.plot(data.month_number, data.moisturizer, color="brown", lw=3, ls='-', marker='o', markersize=6, markerfacecolor="brown")

ax.set_xlabel("Month number")
ax.set_ylabel("Sales units in number")
ax.set_title("Sales data")
plt.legend(("Face cream", "Face wash", "Toothpaste", "Bathing Soap", "Shampoo", "Moisturizer"), loc=2)
```

## Exercise 5
Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.


### The result should look something like this:
<img src="IMG/5.png" width=400>

```python
fig, ax = plt.subplots(figsize=(12,6))
width = 0.25
ax.bar(data.month_number - width/2, data.facecream , width)
ax.bar(data.month_number + width/2, data.facewash, width)
ax.grid(True, linestyle='dashed')
ax.set_xlabel("Month number")
ax.set_ylabel("Sales units in numbers")
ax.set_title("Tooth paste Sales data")
plt.legend(("Face cream sales data","Face wash sales data"), loc=2)
ax.set_xticks(data.month_number)
fig.tight_layout()
```

## Exercise 6
Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

***Note:*** use ``data[item].sum()`` to compute sums over columns 


### The result should look something like this:
<img src="IMG/6.png" width=400>

```python
total = data['total_units'].sum()
total_facecream = (data['facecream'].sum()/total) * 100
total_facewash = (data['facewash'].sum()/total) *100
total_toothpaste = (data['toothpaste'].sum()/total) *100
total_bathingsoap = (data['bathingsoap'].sum()/total) *100
total_shampoo = (data['shampoo'].sum()/total) *100
total_moisturizer = (data['moisturizer'].sum()/total) *100
```

```python
parts_in_percent = [total_facecream, total_facewash, total_toothpaste, total_bathingsoap, total_shampoo, total_moisturizer]
labels = 'Face cream', 'Face wash', 'Toothpaste', 'Bathing soap', 'Shampoo', 'Moisturizer'

fig, ax = plt.subplots(figsize=(12,6))
ax.pie(parts_in_percent, labels=labels, autopct='%1.1f%%',shadow=True)
ax.set_title("Sales data")
plt.legend(("Face cream", "Face wash", "Toothpaste", "Bathing soap", "Shampoo", "Moisturizer"), loc=4)
fig.tight_layout()
```
